```python
import numpy as np 
import time     # Measuring program execution duration
```

```python
# To check wether the python list is faster or numpy array
# we will compare these two with their execution performance by execution time taken.


size=1_000_000     # size 1 million to understand the Process


# Performing square of the elements in the list.
py_list=list(range(size))
start=time.time()       
sq_py_list=[x**2 for x in py_list]
end=time.time()
print(f"Python List Execution time is : {end-start}")


# Performing square of the elements in the numpy-array.
np_arr=np.array(py_list)
start=time.time()
# we will do vectorization ( Most of the bulk functions performed on list using loops are performed without loop in numpy array )
sq_np_arr=np_arr ** 2      # This is Vectorization
end=time.time()
print(f"Numpy-Array Execution time is : {end-start}")
```

```python
# To check wether python list has optimized storage or numpy array (Memory Optimization).
import sys 

print(f"Size of Python List is {sys.getsizeof(py_list)*len(py_list)} bytes")
print(f"Size of Numpy Array is {np_arr.nbytes} bytes")
```
