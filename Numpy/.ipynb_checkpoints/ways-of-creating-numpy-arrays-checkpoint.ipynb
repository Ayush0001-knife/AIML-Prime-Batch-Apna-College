```python
import numpy as np
```

```python
# Create from Lists

arr=np.array([1,2,3,4,5,6])
print(arr,type(arr),arr.dtype,arr.shape)
```

```python
# Create from Lists but with string also there in list

arr=np.array([1,2,3,4,5,6,"Ayush"])
print(arr,type(arr),arr.dtype,arr.shape)   # Every thing inside array will be converted to string 
```

```python
# Create from 2-D Lists 

list=[[1,2,3],[4,5,6],[7,8,9]]
arr2D=np.array(list)
print(arr2D,type(arr2D),arr2D.dtype,arr2D.shape)
```

```python
# Create using Built in methods

arr=np.zeros((2,3))   # Prefill the Array with zeros
print(arr,arr.shape)  # By default it is creatng elemrnts in floating data type

arr1=np.zeros((2,3,),dtype=int)    # we can give the data type also to the array
print(arr1,arr.shape) 
```

```python
# Create using Built in methods

arr=np.ones((2,3))   # Prefill the Array with ones
print(arr,arr.shape)  # By default it is creatng elemrnts in floating data type

arr1=np.ones((2,3,),dtype=int)    # we can give the data type also to the array
print(arr1,arr.shape) 
```

```python
# Create using Built in methods

arr=np.full((2,3),23)   # Prefill the Array with particular number passed after dimension
print(arr,arr.shape)  # By default it is creatng elemrnts in int data type
```

```python
# Create using Built in methods

arr=np.eye((3))   # Creates an Identity Matrix
print(arr,arr.shape)  # By default it is creatng elemrnts in floating data type

arr1=np.eye((3),dtype=int)    # we can give the data type also to the array
print(arr1,arr.shape) 
```
