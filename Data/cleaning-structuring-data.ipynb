```python
import json
```

```python
# Access or Load the Data

def load_data(filename):
    with open (filename,"r") as f:
        data=json.load(f)

    return data

data=load_data("sample_product_rating_data.json")
```

```python
# Clean The Data

def clean_data(data):

    text_to_num = {"one":1,"two":2,"three":3,"four":4,"five":5,}
    cleaned_data=[]
    unique_users=set()
    
    for user in data:
        # Clean the rating because it has mixed data type.
        if "rating" in user:
            # Clean string
            rating = str(user["rating"]).strip().lower()
            
            # Convert text → number
            if rating in text_to_num:
                user["rating"] = text_to_num[rating]

            user["rating"]=rating    

        # Handling missing value.
        raw_age = user.get("age")
        if raw_age is None:
            user["age"] = None

        # Handling Duplicate Data.
        # Deduplication 

        if(user["name"].strip() in unique_users):
            continue
        else:
            unique_users.add(user["name"].strip())
            cleaned_data.append(user)

    return cleaned_data      
```

```python
clean_data(data)
```
